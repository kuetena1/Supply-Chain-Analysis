## Feature Engineering

```python
# let's remove some columns that aren't for the analysis
columns_to_drop = ['Row ID', 'Order ID', 'Product ID', 'Postal Code']
orders_data.drop(columns=columns_to_drop, axis=1, inplace=True)
```

```python
 orders_data.shape
```

```python
orders_data.columns
```

```python
## let change the columns date format and extract the year and month for both the order date and the shipping date
## order date
orders_data['Order Date']= pd.to_datetime(orders_data['Order Date'])
orders_data['year_order'] = orders_data['Order Date'].dt.year
orders_data['month_order'] = orders_data['Order Date'].dt.month_name().str[:3]
## Shipping date
orders_data['Ship Date']= pd.to_datetime(orders_data['Ship Date'])
orders_data['year_ship'] = orders_data['Ship Date'].dt.year
orders_data['month_ship'] = orders_data['Ship Date'].dt.month_name().str[:3]
```

```python
print(orders_data['year_order'].unique())
print(orders_data['year_ship'].unique())
```

```python
## Let's also extract the shipping duration based on the order date and the shipping date
orders_data['shipping_duration'] = (orders_data['Ship Date']- orders_data['Order Date']).dt.days
orders_data.head(2)
```

# Exploratory Analysis

## Sales Performance Analysis 

### Variable-wise Frequency Breakdown: Trends and Patterns

```python

columns_to_plot = ['Ship Mode','Segment', 'Region',
       'Category', 'Sub-Category', 'Returned','year_order', 'month_order']
for column in columns_to_plot:
    plt.figure()
    sns.countplot(data = orders_data, x =column, hue= column)
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.xticks(rotation = 45)
    plt.show()
```

### How Do Product Categories Contribute to Sales and Profit?

```python
## Data
sales_categories =orders_data.groupby('Category')[['Sales', 'Profit']].sum().sort_values(by = 'Sales',ascending= False).reset_index()
# Creating subplots
fig, axes = plt.subplots(1, 2, figsize=(20, 6))

# insights whit categories and sales
sns.barplot(data=sales_categories, x = 'Category', y= 'Sales', hue = 'Category', palette ='bright',ax= axes[0])
axes[0].set_xlabel('category product')
axes[0].set_ylabel('Sales')
axes[0].set_title('Category VS Sales')

# insights with categories  and profit
sns.barplot(data=sales_categories, x = 'Category', y= 'Profit', hue = 'Profit', palette ='Accent' ,ax = axes[1])
axes[1].set_xlabel('category product')
axes[1].set_ylabel('profit')
axes[1].set_title('Category VS Profit')
plt.tight_layout()
plt.show()
```

### How Do Product subcategories Contribute to Sales and Profit for each year?

```python
## Data 
sales_subcategory =orders_data.groupby(['Sub-Category','year_order'])[['Sales', 'Profit']].sum().sort_values(by = 'Sales',ascending= False).reset_index()

# Creating subplots
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
# insights sales with subcategory per year
sns.barplot(data=sales_subcategory, x = 'Sub-Category', y= 'Sales', hue ='year_order', palette ='bright',ax= axes[0])
axes[0].set_xlabel('sales_subcategory product')
axes[0].set_ylabel('Sales')
axes[0].set_title('sales_subcategory VS Sales')
axes[0].tick_params(axis ='x', rotation = 90)

# Profit performance with subcategory per year
sns.barplot(data=sales_subcategory, x = 'Sub-Category', y= 'Profit', hue = 'year_order', palette ='bright' ,ax = axes[1])
axes[1].set_xlabel('sub-category product')
axes[1].set_ylabel('profit')
axes[1].set_title('sales_subcategory VS Profit')
axes[1].tick_params(axis ='x', rotation = 90)
plt.tight_layout()
plt.show()
```

## Insights 

#### Sales Performance
Technology is the top sales and profit driver, indicating strong overall performance. While lower in sales than Technology, Office Supplies manages to achieve the second-highest profit. This suggests a more efficient profit margin compared to Furniture. Furniture, despite being the second-highest in sales, has the lowest profit. This could be due to high costs and lower pricing strategies. It also appears that the subcategory with the highest loss in profit is Tables. followed by Bookcases and supplies across the 4 years we can see that November and December consistently demonstrate the highest sales across all years, particularly 2017 appears to be the strongest year overall, with consistently higher sales than other years in most months. 2014 seems to be the weakest year, showing slow performance in sales. November, December, and February are the most profitable months, especially in 2017 and 2016. march shows high sales but does not translate to equally high profits, indicating potential inefficiencies or higher costs during this period. 2017 not only leads in sales but also generates the highest profits for most months, except for April. 
Standard class shipping mode leads in sales and profit across all years.The Cubify CubeX 3D Printer Double Head Print shows the highest loss. Other notable loss-making products include the Cubify CubeX 3D Printer Triple Head Print and the GBC DocuBind P400 Electric Binding System, which also experienced losses across multiple years. 

#### Shipping performance
the majority of shipping durations are very short (less than 5 days) this indicates that significant portion of orders are either processed and shipped immedialtly or have same-day delivery.  Consumer, Corporate, and Home Office segments have similar shipping durations, with Consumers showing slightly higher variability. The same day" shipping mode has the shortest shipping duration with minimal variation, while the standard Class" has the longest shipping duration with significant variation. West shows the lowest median shipping duration and variability, making it the most consistent region compare to South Central and East. The Envelopes sub-category has a much lower median and range than other sub-categories, indicating faster and more reliable shipping. Sub-categories such as Bookcases, Chairs, and Copiers show a wider range, with more variability and higher shipping durations. 
The plot indicates that shipping durations vary significantly across different months, with some month like February, march, November consistently showing low shipping durations and minimal variability, indicating reliable shipping processes. August stands out with the highest median and widest range of shipping durations .
